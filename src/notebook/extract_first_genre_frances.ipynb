```python
genres = pd.DataFrame(genres)
```

```python
genres['genres'][0]
```

```python
# Split by comma
f = lambda x: x["genres"].split(",")
genres = genres.apply(f, axis=1)
```

```python
genres = pd.DataFrame(genres)
# Extract first genre
f = lambda x: x[0][1]
genres = genres.apply(f, axis=1)
```

```python
genres
```

```python
genres = pd.DataFrame(genres)
# Delete 'name'
f = lambda x: str(x[0]).replace("name", "")
genres = genres.apply(f, axis=1)
```

```python
genres[0]
```

```python
genres = pd.DataFrame(genres)
# Delete '}'
f = lambda x: str(x[0]).replace("}", "")
genres = genres.apply(f, axis=1)
```

```python
genres[0]
```

```python
genres = pd.DataFrame(genres)
# Delete ':'
f = lambda x: str(x[0]).replace(":", "")
genres = genres.apply(f, axis=1)
```

```python
genres = pd.DataFrame(genres)
# Delete ' '
f = lambda x: str(x[0]).replace(" ", "")
genres = genres.apply(f, axis=1)
```

```python
genres = pd.DataFrame(genres)
# Delete ']'
f = lambda x: str(x[0]).replace("]", "")
genres = genres.apply(f, axis=1)
```

```python
genres = pd.DataFrame(genres)
# Delete "'"
f = lambda x: str(x[0]).replace("'", "")
genres = genres.apply(f, axis=1)
```
