```python
# Import csv
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/Users/FrancesY/Documents/Columbia/spring2022/5063viz/project/data/cleaned4.csv")
df.drop(columns = "Unnamed: 0", inplace = True)
```

```python
df
```

```python
sns.set(rc = {'figure.figsize':(25,25)})

p = sns.violinplot(x="budget", y="genres",
                    data=df,
                    cut=0, scale="width", palette="Set3")
sns.set(font_scale = 4)
p.set_xlabel("Budget($100 Million)", fontsize=30)
p.set_ylabel("Movie Genres", fontsize=30)
p.set(title='Budget By Movie Genres')
plt.xticks(fontsize = 25)
plt.yticks(fontsize = 25) 
plt.savefig('budget-by-genres.png', dpi=300)
```

```python
sns.set(rc = {'figure.figsize':(25,25)})

p = sns.violinplot(x="revenue", y="genres",
                    data=df,
                    cut=0, scale="width", palette="Set3")
sns.set(font_scale = 4)
p.set_xlabel("Revenue($1000 Million)", fontsize=30)
p.set_ylabel("Movie Genres", fontsize=30)
p.set(title='Revenue By Movie Genres')
plt.xticks(fontsize = 25)
plt.yticks(fontsize = 25) 
plt.savefig('revenue-by-genres.png', dpi=300)
```
