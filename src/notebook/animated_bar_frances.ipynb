```python
# Count for each genre
genres_count = df['genres'].value_counts()
genres_count = pd.DataFrame(genres_count)
```

```python
genres_10 = genres_count.head(10)
```

```python
genres_tail = genres_count.tail(10)
genres_tail = pd.DataFrame(genres_tail)
```

```python
# Add 'other' category
genres_10.loc['Other'] = genres_tail['genres'].sum()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(8, 6), dpi=80)

colors = ['#E8BDC4', '#B7C3F3', '#DD7596', '#8EB897', '#D5D6EA', '#F6F6EB', '#D7ECD9', '#F5D5CB','#F6ECF5','#F3DDF2','#C0DEAD']

plt.pie(genres_10['genres'],
        labels=genres_10.index, 
        labeldistance=1.15, 
        startangle=90, 
        shadow=True, 
        autopct='%1.1f%%',
        colors = colors)

plt.title('Movie Genres')

plt.savefig('genre_pie.png')

plt.show()

# This pie chart shows the percentage of each movie genre produced of all time.
```

```python
# Group by year and genre
sizes = df.groupby(['release_year', 'genres']).size()
sizes = sizes.reset_index()
sizes = sizes.rename(columns = {0:"Count"})
```

```python
sizes
```

```python
sizes2 = pd.pivot_table(sizes,index='release_year',columns=['genres'],values='Count')
```

```python
sizes2
```

```python
sizes2 = sizes2.stack(level=[0],dropna=False).to_frame('Count').reset_index()
```

```python
sizes2
```

```python
sizes2=sizes2.fillna(0)
```

```python
sizes2
```

```python
# Add cumulative count
sizes2['Cumulative'] = sizes2.groupby("genres")['Count'].cumsum()
```

```python
sizes2
```

```python
for col in sizes2.release_year.values:
...     sizes2['release_year'] = sizes2['release_year'].astype('int64')
```

```python
sizes2
```

```python
sizes2['Cumulative'].max()
```

```python
import plotly.express as px
fig = px.bar(sizes2, x="genres", y="Cumulative", color="genres", color_discrete_sequence=px.colors.qualitative.G10,
animation_frame="release_year", animation_group="genres", range_y=[0,1000])

fig.update_layout(
                  title="Movie Genre Changes Overtime",
    yaxis_title = "Cumulative Count",
    font_family="Courier New",
    font_color="#013220",
    legend_title = "Movie Genre",
    legend_title_font_color="#023020"
)

fig.update_layout(xaxis={'title': '',
                        'visible': True,
                        'showticklabels': True})

fig.write_html("genre_overtime_plotly.html")

fig.show()
```
