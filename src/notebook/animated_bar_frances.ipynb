```python
# Import csv
import pandas as pd
import numpy as np

df = pd.read_csv("/Users/FrancesY/Documents/Columbia/spring2022/5063viz/project/data/cleaned3.csv")
df.drop(columns = "Unnamed: 0", inplace = True)
```

```python
df
```

```python
# Count for each genre
genres_count = df['genres'].value_counts()
genres_count = pd.DataFrame(genres_count)
```

```python
genres_10 = genres_count.head(10)
```

```python
genres_tail = genres_count.tail(10)
genres_tail = pd.DataFrame(genres_tail)
```

```python
# Add 'other' category
genres_10.loc['Other'] = genres_tail['genres'].sum()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(8, 6), dpi=80)

colors = ['#E8BDC4', '#B7C3F3', '#DD7596', '#8EB897', '#D5D6EA', '#F6F6EB', '#D7ECD9', '#F5D5CB','#F6ECF5','#F3DDF2','#C0DEAD']

plt.pie(genres_10['genres'],
        labels=genres_10.index, 
        labeldistance=1.15, 
        startangle=90, 
        shadow=True, 
        autopct='%1.1f%%',
        colors = colors)

plt.title('Movie Genres')

plt.savefig('genre_pie.png')

plt.show()

# This pie chart shows the percentage of each movie genre produced of all time.
```

```python
# Group by year and genre
sizes = df.groupby(['release_year', 'genres']).size()
sizes = sizes.reset_index()
sizes = sizes.rename(columns = {0:"Count"})
```

```python
for col in sizes.release_year.values:
...     sizes['release_year'] = sizes['release_year'].astype('int64')
```

```python
sizes
```

```python
sizes = sizes[sizes['release_year'] > 1980]
sizes
```

```python
sizes['Count'].max()
```

```python
import plotly.express as px
fig = px.bar(sizes, x="genres", y="Count", color="genres", color_discrete_sequence=px.colors.qualitative.G10,
animation_frame="release_year", animation_group="genres", range_y=[0,300])

fig.update_layout(title="Movie Genre Changes Overtime",
    font_family="Courier New",
    font_color="#013220",
    legend_title_font_color="#023020"
)

fig.write_html("genre_overtime_plotly.html")

fig.show()
```
