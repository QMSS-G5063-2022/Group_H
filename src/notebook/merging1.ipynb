```python
import pandas as pd
```

```python
keywords = pd.read_csv("/Users/FrancesY/Documents/Columbia/spring2022/5063viz/project/dataset//keywords.csv")
```

```python
movies = pd.read_csv("/Users/FrancesY/Documents/Columbia/spring2022/5063viz/project/dataset/movies_metadata.csv")
```

```python
movies2 = movies[movies['original_language'] == 'en']
```

```python
credits = pd.read_csv("/Users/FrancesY/Documents/Columbia/spring2022/5063viz/project/dataset/credits.csv")
```
