```python
# Import csv
import pandas as pd
import numpy as np

df = pd.read_csv("/Users/FrancesY/Documents/Columbia/spring2022/5063viz/project/data/cleaned4.csv")
df.drop(columns = "Unnamed: 0", inplace = True)
```

```python
df
```

```python
import plotly.express as px

# Scatter plot of popularity vs. revenue


fig = px.scatter(df, x="vote_average", y="revenue", color="genres", size='popularity',hover_name='title',size_max=35)



fig.update_layout(title="Vote Average vs. Revenue",
    xaxis_title = "Vote Average",
    font_family="Courier New",
    font_color="#013220",
    legend_title_font_color="#023020"
)

fig.write_html("vote_revenue_plotly_frances.html")


fig.show()
```
